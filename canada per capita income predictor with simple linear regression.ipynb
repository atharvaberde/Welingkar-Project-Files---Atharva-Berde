```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("C:/Users/atharva/OneDrive/Desktop/self learn data science/machine learning algorithms/simple linear/canada per capital income prediction.csv")
df
```

```python
%matplotlib inline

plt.xlabel('year')
plt.ylabel('pci')
plt.scatter(df.year,df.pci, color = 'red', marker = '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.pci)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
input_value = 2020
input_value_reshaped = np.array([[input_value]]) 
prediction = reg.predict(input_value_reshaped)
prediction
```
