```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df= pd.read_csv("C:/Users/atharva/OneDrive/Desktop/self learn data science/machine learning algorithms/logistic regression/flower petal prediction/IRIS.csv")
df
```

```python
#check data of DV
df['species'].unique()
```

```python
#replace the species value with numeric values
df['species'] = df['species'].replace({'Iris-setosa':1,'Iris-versicolor':2,'Iris-virginica':3})
```

```python
df
```

```python
x_train,x_test,y_train,y_test = train_test_split(df[['sepal_length','sepal_width','petal_length','petal_width']],df['species'],train_size=0.8)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn import linear_model
model= linear_model.LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
x_test
```

```python
predicted_output = model.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
com = confusion_matrix (y_test, predicted_output)
com
```
