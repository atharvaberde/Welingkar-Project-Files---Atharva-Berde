```python
import pandas as pd
df = pd.read_csv("C:/Users/atharva/OneDrive/Desktop/self learn data science/machine learning algorithms/Decision Tree Classificaion/PlayTennis.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
outlook = LabelEncoder()
temp = LabelEncoder()
humidity = LabelEncoder()
windy = LabelEncoder()
play = LabelEncoder()
```

```python
df['outlook']=outlook.fit_transform(df['outlook'])
df['humidity']=outlook.fit_transform(df['outlook'])
df['windy']=outlook.fit_transform(df['outlook'])
df['temp']=outlook.fit_transform(df['outlook'])
df['play']=outlook.fit_transform(df['outlook'])
```

```python

features_cols = ['outlook', 'humidity', 'windy', 'temp']
x = df[features_cols]
y = df.play
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier(criterion = 'gini')
```

```python
classifier.fit(x_train, y_train)
```

```python
classifier.predict(x_test)
```

```python
x_test
```

```python
classifier.score(x_test,y_test)
```
